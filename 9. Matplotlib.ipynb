# Matplotlib

Matplotlib is a popular and powerful library used for creating static, interactive, and animated visualizations. It provides a wide range of functions to generate high-quality 2D and 3D plots.

```python
import matplotlib.pyplot as plt
```

```python
year = [1950,1970,1990,2010]
pop = [2.519,3.692,5.263,9.972]
plt.plot(year,pop)
plt.show()
```

### Scatter plot

```python
plt.scatter(year,pop)
plt.show()
```

### Histogram

Graphical representation of the distribution of a dataset.The x-axis represents the bins, and the y-axis represents the frequency of values falling into each bin. In Python, if you do not specify, by default, bins=10.

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 6, 6, 7, 8, 8, 8, 9, 10]

plt.hist(data, bins=5)
#plt.clf() > clear figure

plt.show()
```

#### Label axes and Title

```python
year = [1950,1970,1990,2010]
pop = [2.519,3.692,5.263,9.972]
plt.plot(year,pop)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title("World Population")

plt.show()
```

```python
year = [1950,1970,1990,2010]
pop = [2.519,3.692,5.263,9.972]
plt.plot(year,pop)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title("World Population")

## Change the labels

plt.yticks([0,2,4,6,8,10],['0', '2B', '4B','6B','8B','10B'])

plt.show()
```

```python
import numpy as np

year = [1950,1970,1990,2010]
pop = [2.519,3.692,5.263,9.972]

np_pop = np.array(pop)
np_pop =  np_pop*20

plt.scatter(year,pop, s = np_pop)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title("World Population")

## Change the labels

plt.yticks([0,2,4,6,8,10],['0', '2B', '4B','6B','8B','10B'])

plt.show()
```

```python
year = [1950,1970,1990,2010]
pop = [2.519,3.692,5.263,9.972]

## adding different colors
col = ['red','green','blue','yellow']

## alpha = transparency
plt.scatter(x = year, y = pop, s = np.array(pop) * 20, alpha = 0.4, c = col)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title("World Population")

plt.yticks([0,2,4,6,8,10],['0', '2B', '4B','6B','8B','10B'])

plt.show()
```

```python
year = [1950,1970,1990,2010]
pop = [2.519,3.692,5.263,9.972]

## adding different colors
col = ['red','green','blue','yellow']

## alpha = transparency
plt.scatter(x = year, y = pop, s = np.array(pop) * 20, alpha = 0.4, c = col)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title("World Population")
plt.yticks([0,2,4,6,8,10],['0', '2B', '4B','6B','8B','10B'])

## adding text
plt.text(1950,3,'2010')
plt.text(2005,9,'2010')

## adding grid
plt.grid(True)
plt.show()
```
